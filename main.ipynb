## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    #Le damos un valor a factorial de 1
    fact=1
    #mientras el numero del que estamos calculando su valor factorial sea mayor que uno:
    while n>1:
        #la variable factorial va cogiendo el valor del n¬∫ multiplicado por el mismo n¬∫ menos 1 y vuelve a restar uno al n¬∫ y a multiplicar
        #as√≠ hasta que el n¬∫ no es mayor que 1 y el loop se rompe
        fact*=n
        n-=1
              
    return fact
```

```python
# This will test your function 
test_factorial(factorial)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. üòâ`

```python
from statistics import stdev
def st_dev(arr):
    #para llegar a la standard deviation primero hay que enecontrar el mean de la lista
    #para ello hay que sumar todos los elementos de la lista y dividirlo por el numero de elemtos en ella
    #creamos la variable suma y n¬∫ de elementos y le damos valor cero
    suma=0
    numero_de_elementos=0
    #iteramos por la lista sumando todos los elemetos y cada vez que itera sumamos un elemento a la variable de n¬∫ de elementos
    for i in arr:
        suma+=i
        numero_de_elementos+=1
    #dividimos la suma entre el total de elemtos  ya tenemos el mean
    mean=suma/numero_de_elementos

    #lo siguiente que hay que hacer es extraer el mean de cada uno de los elementos y calcular el cuadrado de cada uno
    lista_mean_subtract=[]
    for i in arr:
        lista_mean_subtract.append((i-mean)**2)
    
    #el siguiente paso es sumar todos los elemntos de la lista
    suma2=0
    for i in lista_mean_subtract:
        suma2+=i

    #La suma de los cuadrados, lo dividimos entre el n¬∫ de elemntos de la lista -1
    standard_sample_deviation=suma2/(numero_de_elementos-1)
    #y finalmente sacamos la ra√≠z cuadrada del n¬∫
    standard_dev_final=standard_sample_deviation**0.5
    
    return (standard_dev_final)
```

```python
# This will test your function 
test_stdev(st_dev)
```
